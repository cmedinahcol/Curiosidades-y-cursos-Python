Objetivo: Se quiere hayar a los estudiantes que no armaron grupos de trabajo pero que están inscritos en el curso:

Conclusiones:


* En resultado2 se ve que hay valores raros... posiblemente los estudiantes escribieron mal su correo o hay valores vacios.


* En resultado3 se encuentra la correcta escritura de los estudiantes resultado2 y se hayan los estudiantes que no armaron grupos de trabajo.

# Codigo total resumido:

```python
#Cargar libreria y base de datos:

import pandas as pd
estudiantes=pd.read_excel('C:/Users/David/Downloads/FF122 LISTA ESTUDIANTES 20-03-22.xlsx')

#organizar datos:

personas_totales=pd.Series(estudiantes['personas totales'].sort_values(ascending=True)).reset_index()
personas_totales=personas_totales['personas totales'] #Serie personas completas del curso
grupos=pd.Series(estudiantes['grupos']) #Serie personas anotadas en grupos

# ^ Realiza bit a bit la operación XOR en los operandos. Es decir, muestra como salida los valores que no coinciden:

resultado=set(personas_totales)^set(grupos)

#Mostrar a que serie pertenecen los valores que no coinciden:
# & va bit a bit y muestra los valores que coinciden:

resultado2=set(resultado)&set(grupos) 
resultado3=set(resultado)&set(personas_totales) 

print(resultado2) #Valores raros en personas anotadas en grupos
print(resultado3) #Valores raros en personas completas del curso
```
